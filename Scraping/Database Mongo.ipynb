```python
import csv
import json
import pandas as pd
import sys, getopt, pprint
from pymongo import MongoClient

#import sneakersnstuff
csvfile = open('/Users/fabienbarrios/Documents/PROJET NOSQL/Script/sneakersnstuff_produit.csv', 'r')
reader = csv.DictReader(csvfile)
client = MongoClient('localhost',27017)
db = client.sneakers
db.sneakersnstuff.drop()
header= ["Brand", "Model", "Price"]

for each in reader:
    row={}
    for field in header:
        row[field]=each[field]
    
    db.sneakersnstuff.insert(row)
```
